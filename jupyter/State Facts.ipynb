```python
import pandas as pd
import re

from amz_local_search.resources.fba_centers import fba_centers_info
from amz_local_search.resources.states.us_state_abbrev import us_state_abbrev
```

```python
# List taken from https://en.wikipedia.org/wiki/List_of_largest_cities_of_U.S._states_and_territories_by_population
wiki_pop_df = pd.read_csv('/Users/evgenyp/Projects/AmzJet/AmzJet/amz-apps/amz_local_search/resources/references/State Cities by Population.tsv', sep='\t')
wiki_pop_df.head(3)
```

```python
wiki_pop_df['State'] = wiki_pop_df['State, Federal District, or Territory'].map(lambda s: us_state_abbrev.get(s))
```

```python
wiki_pop_df_cleaned = wiki_pop_df[pd.notnull(wiki_pop_df['State'])][['State', 'State, Federal District, or Territory Population[5]', 'Most populous']]
wiki_pop_df_cleaned.columns = ['state', 'population', 'major_city']
wiki_pop_df_cleaned['major_city'] = wiki_pop_df_cleaned['major_city'].map(lambda n: re.split('\d', n)[0])
wiki_pop_df_cleaned['population'] = wiki_pop_df_cleaned['population'].map(lambda s: s.replace(',', '')).astype(int)

wiki_pop_df_cleaned['num_fba_codes'] = wiki_pop_df_cleaned['state'].map(lambda s: len(fba_centers_info.state_to_codes(s)))

#wiki_pop_df_cleaned.set_index('state', inplace=True)
wiki_pop_df_cleaned.to_csv('/Users/evgenyp/Projects/AmzJet/AmzJet/amz-apps/amz_local_search/resources/states/facts.tsv', sep='\t')
wiki_pop_df_cleaned.to_dict(orient='records')
```
